# Access Census data
Steps:
1. Check which data and which API to use: (http://www.census.gov/data/developers/data-sets.html) In this example, I am using ACS 2009-2013 5-year estimate (http://www.census.gov/data/developers/data-sets/acs-survey-5-year-data.html)
2. Check example codes and geography: (http://api.census.gov/data/2013/acs5/examples.html)
3. Figure out the example codes, for example (http://api.census.gov/data/2013/acs5?get=NAME,B01001_001E&for=block+group:0&in=state:06+county:061+tract:990000&key=26cb4e7274c2a71e9176eda735a44d192dbd6c62)
4. Figure out variables
5. Rewrite URL
6. Request

```python
import requests as re
import pandas as pd

basicURL = 'http://api.census.gov/data/2013/acs5?get='
table = 'NAME,B01001_001E,B06011_001E'
geo = 'for=block+group:*&in=state:24+county:031+tract:*'
api_key = 'key=26cb4e7274c2a71e9176eda735a44d192dbd6c62'

url = basicURL + table + '&' + geo + '&' + api_key
r = re.get(url)

print r
```

```python
request_df = pd.DataFrame(r.json())
request_df
```

```python
request_df.columns = request_df.iloc[0]
request_df = request_df[1:]
request_df
```
